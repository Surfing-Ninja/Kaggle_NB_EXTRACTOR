```python
import pandas as pd
```

```python
A = pd.read_csv('/kaggle/input/gaf-submission/submission.tsv',sep='\t',header=None)
A['pred_key'] = [f'{pt}_{go}' for pt,go in zip(A[0],A[1])]
A.tail()
```

```python
B = pd.read_csv('/kaggle/input/merge-of-2submission-lb-0-25/submission.tsv',sep='\t',header=None)
B['pred_key'] = [f'{pt}_{go}' for pt,go in zip(B[0],B[1])]
B.tail()
```

```python
common_in_A = A[A['pred_key'].isin(B['pred_key'])].sort_values('pred_key').reset_index(drop=True)
common_in_A.tail()
```

```python
common_in_B = B[B['pred_key'].isin(A['pred_key'])].sort_values('pred_key').reset_index(drop=True)
common_in_B.tail()
```

```python
common_ensemble = common_in_A.copy()
common_ensemble[2] = (common_in_A[2] + common_in_B[2]) / 2
```

```python
not_common_in_A = A[~A['pred_key'].isin(B['pred_key'])].sort_values('pred_key').reset_index(drop=True)
not_common_in_A.tail()
```

```python
not_common_in_B = B[~B['pred_key'].isin(A['pred_key'])].sort_values('pred_key').reset_index(drop=True)
not_common_in_B.tail()
```

```python
pd.concat([
    common_ensemble,
    not_common_in_A,
    not_common_in_B
])[[0,1,2]].to_csv('submission.tsv',sep='\t',index=False,header=None)
```
