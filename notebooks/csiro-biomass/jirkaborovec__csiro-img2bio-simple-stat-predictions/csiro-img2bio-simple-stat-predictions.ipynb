```python
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
PATH_DATA = "/kaggle/input/csiro-biomass"
```

## Browse the metadata

```python
# You might need to adjust the file path based on the actual extracted files.
test_df = pd.read_csv(f'{PATH_DATA}/test.csv')
display(test_df.head())
```

```python
# You might need to adjust the file path based on the actual extracted files.
train_df = pd.read_csv(f'{PATH_DATA}/train.csv')
display(train_df.head())
```

## Dictribution

Plot the distribution of the `target` column for each `target_name` class in the `train` table.

```python
import matplotlib.pyplot as plt
import seaborn as sns

grouped_train_data = train_df.groupby('target_name')

plt.figure(figsize=(12, 8))
for target_name, group_data in grouped_train_data:
    sns.histplot(data=group_data, x='target', kde=True, label=target_name)

plt.title('Distribution of Target for Each Target Name Class')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.legend(title='Target Name')
plt.grid(True) # Added grid here
plt.show()
```

```python
target_median_lookup = train_df.groupby('target_name')['target'].median().to_dict()
print(target_median_lookup)
```

## Stat prediction

```python
test_df['target'] = test_df['target_name'].map(target_median_lookup)
display(test_df.head())
```

```python
# Create a submission dataframe with 'sample_id' and 'target' columns
submission_df = test_df[['sample_id', 'target']]

# Save the submission dataframe to a CSV file
submission_df.to_csv('submission.csv', index=False)

print("Submission file 'submission.csv' created successfully.")
!head submission.csv
```
