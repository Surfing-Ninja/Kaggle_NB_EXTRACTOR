```python
import numpy as np 
import pandas as pd 
```

```python
# Load the train dataset
df_train = pd.read_csv('/kaggle/input/csiro-biomass/train.csv')

print(df_train.shape)
df_train.head()
```

```python
# Calculate the median of 'target' grouped by 'target_name'
mean_target_by_name = df_train.groupby('target_name')['target'].median().reset_index()
mean_target_by_name
```

```python
# Load the test dataset
df_test = pd.read_csv('/kaggle/input/csiro-biomass/test.csv')

print(df_test.shape)
df_test.head()
```

```python
submission_df = df_test.merge(
    mean_target_by_name,
    on='target_name',
    how='left'
)[["sample_id","target"]]

submission_df.head()
```

```python
submission_df.to_csv('submission.csv', index=False)
```
