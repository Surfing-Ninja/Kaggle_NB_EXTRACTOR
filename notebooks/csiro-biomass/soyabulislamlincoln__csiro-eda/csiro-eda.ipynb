```python
def plot_multiple_images_per_class(df, image_path_col, class_col, samples_per_class=3):
  
    classes = df[class_col].unique()
    
    # Create subplot grid
    fig, axes = plt.subplots(len(classes), samples_per_class, 
                            figsize=(15, 5*len(classes)))
    
    # Handle single class case
    if len(classes) == 1:
        axes = axes.reshape(1, -1)
    
    for row, class_name in enumerate(classes):
        # Get images for this class
        class_images = df[df[class_col] == class_name][image_path_col].tolist()
        
        # Take first 'samples_per_class' images
        selected_images = class_images[:samples_per_class]
        
        for col, img_path in enumerate(selected_images):
            ax = axes[row, col] if len(classes) > 1 else axes[col]
            
            # Read and process image
            if os.path.exists(img_path):
                img = cv2.imread(img_path)
                if img is not None:
                    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    
                    # Display image
                    ax.imshow(img_rgb)
                    ax.set_title(f'Class: {class_name}\n{os.path.basename(img_path)}', 
                                fontsize=10, pad=5)
                    ax.axis('off')
                else:
                    # Display placeholder if image couldn't be loaded
                    ax.text(0.5, 0.5, 'Image\nnot found', 
                           ha='center', va='center', transform=ax.transAxes,
                           fontsize=12, color='red')
                    ax.axis('off')
            else:
                # Display placeholder if file doesn't exist
                ax.text(0.5, 0.5, 'File\nnot found', 
                       ha='center', va='center', transform=ax.transAxes,
                       fontsize=12, color='red')
                ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python

base_dir = "/kaggle/input/csiro-biomass"
train_data['full_image_path'] = train_data['image_path'].apply(lambda x: os.path.join(base_dir, x))


plot_multiple_images_per_class(train_data, 'full_image_path', 'Species', samples_per_class=3)
```

```python
plot_multiple_images_per_class(train_data, 'full_image_path', 'target_name', samples_per_class=3)
```
