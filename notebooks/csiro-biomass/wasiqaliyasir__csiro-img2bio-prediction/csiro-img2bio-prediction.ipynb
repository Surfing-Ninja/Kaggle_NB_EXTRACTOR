```python
import numpy as np
import pandas as pd
import torch

# ====================================================
# Configuration
# ====================================================
class CFG:
    output_file = "submission.csv"   # Final output in CSV format
    num_samples = 5                  # Number of predictions to generate
    model_name = "convnext_tiny"
    folds = 5
    device = "cuda" if torch.cuda.is_available() else "cpu"

# ====================================================
# Dummy Prediction Generator (replace with real model preds if available)
# ====================================================
def generate_predictions(num_samples, folds):
    """
    Generate dummy predictions for submission.
    Replace with real model predictions if available.
    """
    preds = []
    for fold in range(folds):
        print(f"âœ… Simulating fold {fold} predictions...")
        fold_preds = np.random.rand(num_samples) * 100  # Example predictions
        preds.append(fold_preds)
    final_preds = np.mean(preds, axis=0)
    return final_preds

# ====================================================
# Inference Function
# ====================================================
def run_inference():
    print("ðŸš€ Starting Inference...")

    # Simulate model inference (replace with real model outputs)
    final_preds = generate_predictions(CFG.num_samples, CFG.folds)

    # Create simple DataFrame for submission
    sub_df = pd.DataFrame({
        "sample_id": [f"sample_{i}" for i in range(len(final_preds))],
        "biomass": final_preds
    })

    # Save CSV file
    sub_df.to_csv(CFG.output_file, index=False)
    print(f"\nâœ… Submission file saved successfully: {CFG.output_file}")
    print(sub_df.head())

# ====================================================
# Run
# ====================================================
if __name__ == "__main__":
    run_inference()
```
