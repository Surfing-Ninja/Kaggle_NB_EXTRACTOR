# 比赛描述

<font size=3> &emsp;&emsp;MNIST（Modified National Institute of Standards and Technology）是美国国家标准与技术研究院收集整理的大型手写数字数据库。自 1999 年发布以来，这个经典的手写图像数据集一直作为基准分类算法的基础。随着新的机器学习技术的出现，MNIST 仍然是研究人员和学习者的可靠资源。 \
    &emsp;&emsp;在本次比赛中，您的目标是从数万张手写图像的数据集中正确识别数字。我们鼓励尝试不同的算法，以亲身了解哪些算法行之有效以及技术如何比较。</font>

# 数据及具体目标

<font size=3> &emsp;&emsp;**数据包括三个文件：train.csv, test.csv, sample_submission.csv** \
&emsp;&emsp;其中，train.csv 和 test.csv 包含从0到9的手绘数字的灰度图像。每张图像的高度为 28 像素，宽度为 28 像素，总共为 784 像素。每个像素都有一个与之关联的像素值，指示该像素的亮度或暗度，**数字越大表示越暗**。此像素值是介于 0 和 255 之间的整数（包括 0 和 255）。\
&emsp;&emsp;训练集 train.csv 有 785 列。第一列是标签（label），表示用户绘制的数字；其余列是关联图像的像素值。训练集中的每个像素列都有一个类似于 pixelx 的名称，其中 x 是介于 0 和 783 之间的整数（包括 0 和 783）。为了在图像上找到这个像素，假设我们将 x 分解为 x = i * 28 + j，其中i和j是0和27之间的整数，包括0和27。则 pixelx 位于 28 x 28 矩阵的行 i 和列 j 上（按零索引）。例如，pixel31 表示位于第四列第二行的像素。 \
&emsp;&emsp;从视觉上看，如果我们省略“像素”前缀，像素将组成图像，如下所示：\
&emsp;&emsp;&emsp;&emsp;000 001 002 003 ... 026 027 \
&emsp;&emsp;&emsp;&emsp;028 029 030 031 ... 054 055 \
&emsp;&emsp;&emsp;&emsp;056 057 058 059 ... 082 083 \
&emsp;&emsp;&emsp;&emsp;............................................. \
&emsp;&emsp;&emsp;&emsp;728 729 730 731 ... 754 755 \
&emsp;&emsp;&emsp;&emsp;756 757 758 759 ... 782 783 \
&emsp;&emsp;测试集 test.csv 与训练集相同，只是它不包含“标签”列。\
&emsp;&emsp;提交文件应采用以下格式：对于测试集中的 28000 张图像中的每张，输出一行，其中包含 ImageId 和您预测的数字。例如，如果您预测第一个图像为 3，第二个图像为 7，第三个图像为 8，则提交文件将如下所示：  \
&emsp;&emsp;&emsp;&emsp;ImageId,Label     \
&emsp;&emsp;&emsp;&emsp;1,3            \
&emsp;&emsp;&emsp;&emsp;2,7            \
&emsp;&emsp;&emsp;&emsp;3,8            \
&emsp;&emsp;&emsp;&emsp;(27997 more lines)  \
&emsp;&emsp;**最终目标：对于测试集中的 28000 张图像，预测每张图像所对应的数字。**   
&emsp;&emsp;**评估指标：分类准确性，也就是正确分类测试图像的比例。例如，分类准确率为 0.97 表示正确分类除 3% 之外的所有图像。**

### 归一化（Normalization）

我们执行灰度归一化以减少照明差异的影响。 \
此外，CNN 在 [0, 1] 数据上的收敛速度比在 [0, 255] 上的收敛速度更快。

```python
X_train = X_train / 255.0
test = test / 255.0
```

### 重塑图像（Reshape）

&emsp;&emsp;注意到训练和测试图像 (28px × 28px) 是作为 包含 784 个值的一维向量存入数据框的，结合上述介绍的背景知识，我们以 3 个维度重塑图像（高度 = 28px，宽度 = 28px，通道 = 1），将所有数据重塑为 28x28x1 3D 矩阵。 \
&emsp;&emsp;这里多了个通道维度，是因为后续建模会用到 Keras 。 Keras 最终需要一个额外的维度来对应通道。
MNIST 图像是灰度的，它只使用一个通道。 对于 RGB 图像，有 3 个通道，我们会将 784px 向量重新整形为 28x28x3 3D 矩阵。

```python
# 以 3 个维度重塑图像(height = 28px, width = 28px , canal = 1)
X_train = X_train.values.reshape(-1, 28, 28, 1)
test = test.values.reshape(-1, 28, 28, 1)
```

## 卷积神经网络（CNN）
### 定义模型

&emsp;&emsp;[卷积神经网络（Convolutional Neural Networks, CNN）][1]是深度学习中非常常见的模型，其在图像处理中应用广泛。那么，对于其原理和相关概念在这里就不做[详细的介绍][2]了,下面我主要讲一下模型的架构是怎样的。

[1]: https://blog.csdn.net/u010555688/article/details/24487301
[2]: https://zhuanlan.zhihu.com/p/344562609

<font size=3>
    
1. 首先，我们输入 28*28 像素的单通道图像。 
2. 其次，通过两层卷积和池化操作，使得 CNN 能够结合局部特征并学习图像的更多全局特征。
- 第一层：第一个图是卷积层，使用 5* 5 的32个**卷积核（Conv2D）**，它好比一组可学习的过滤器，将图像进行线性变换，并通过**激活函数 ReLU** ，将这个卷积层的线性计算系统引入非线性特征。第二个图是池化层，进行 2* 2 的**最大值池化（MaxPool2D）**，它充当一个下采样过滤器，查看 2 个相邻像素并选择最大值，将原始 28* 28 像素的图像缩小为 14* 14 像素的图像，这能降低计算成本，并在一定程度上减少过拟合。

- 第二层：同样第一个图使用 5* 5 的64个卷积核，并通过激活函数 ReLU 进行非线性变换。第二个图进行 2* 2 的最大值池化（MaxPool2D），将 14* 14 像素的图像进一步缩小为 7* 7 像素的图像。
3. 然后，通过 Flatten（展平）将最终的特征图转换为一个一维向量，也就是第一个完全连接层中的 7*7*64=3164 个节点，那么它结合了之前卷积层的所有找到的局部特征3164个。
4. 最后，第二个完全连接层（也就相当于人工神经网络ANN的隐藏层）包含1000个节点，第一层到第二层通过ReLU激活函数；最后我们通过Softmax激活函数输出10个节点，也就是10种类别数字标签的概率分布。
    
&emsp;&emsp;**注：** 训练模型之后，神经网络的权重与训练样本太过匹配，往往会导致在处理新样本的时候表现平平。Dropout 是一种正则化方法，对于每个训练样本，它会将层中的一部分节点随机丢弃掉（即将它们的权重 wieght 设置为零）。这会迫使神经网络以分布式的方法去学习特征，从而提高模型的泛化能力并减少过拟合。
    
</font>

```python
# 设置CNN模型，我的CNN架构是 -> [Conv2D->relu -> MaxPool2D -> Dropout]*2 -> Flatten -> Dense -> Dropout -> Out
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', activation ='relu', input_shape = (28,28,1)))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 64, kernel_size = (5,5),padding = 'Same', activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(1000, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))
```

### 设置优化器和退火器

<font size=3>
    
定义好模型之后，我们需要设置**优化器、损失函数和评价函数以及学习率退火器**。

- 首先设置优化器，优化器能迭代地改进参数（过滤核值、权重和神经元的偏差等等）以最小化损失。常见的优化器有随机梯度下降（sgd）、Adam和RMSprop，这里我们选择RMSprop。RMSprop 是一个十分有效的优化器，并且比随机梯度下降要快，它是一种 AdaGrad 调整方法，通过把累计梯度加权平均，试图降低其激进、单调递减的学习率。
    

- 其次，选取损失函数。损失函数用来衡量模型在具有已知标签的图像上的表现有多差，具体指的是观察到的标签和预测的标签之间的错误率。我们使用“categorical_crossentropy”的多分类损失函数，而“binary_crossentropy”一般用于二分类。
    

- 然后，选取评价函数“准确度”用于评估我们模型的性能。即题目要求的分类准确性，用 accuracy 表示。
    

- 最后，为了使优化器更快收敛并最接近损失函数的全局最小值，还需要设置学习率退火器。学习率越高，步数越大，收敛越快。然而，高学习率的采样非常差，优化器可能会陷入局部最小值。最好是在训练期间降低学习率，以有效地达到损失函数的全局最小值。为了保持高学习率的快速计算时间的优势，每 X 步（时期）动态降低学习率，这取决于是否有必要（当精度没有提高时）。为了更好的调节学习率，可以使用 keras.callbacks 中的 ReduceLROnPlateau 回调函数，它会随着训练而自动调节学习率，如果 3 个 epoch 后精度没有提高，我选择将学习率减少一半。
    
</font>

```python
# 设置优化器并编译模型
optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
model.compile(optimizer = optimizer, loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
# 设置学习率退火器
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy',
                                            factor=0.5,
                                            patience=3,
                                            verbose=1,
                                            min_lr=0.00001)
```

### 数据增强（Data augmentation ）

&emsp;&emsp;为了避免过拟合问题，需要人为地扩展我们的手写数字数据集。以改变数组表示同时保持标签相同的方式改变训练数据的方法被称为数据增强技术。
我们可以通过灰度、水平翻转、垂直翻转、随机裁剪、颜色抖动、平移、旋转等操作来进行数据的增强，使得现有的数据集更大，同时达到防止过拟合，提高泛化能力的作用。

对于数据增强，这里选择： \
&emsp;&emsp;（1）将训练图像随机旋转 10 度 \
&emsp;&emsp;（2）将训练图像随机缩放 10% \
&emsp;&emsp;（3）将图像水平随机移动 10% 的宽度 \
&emsp;&emsp;（4）将图像垂直随机移动 10% 的高度 

注：这里没有应用vertical_flip 或horizontal_flip，因为水平翻转或垂直翻转会导致错误分类对称数字，例如 6 和 9。

改进很重要：
  - 在没有数据增强的情况下，准确率为 98.114% ；
  - 通过数据增强，准确率可以达到 98.7+% 。

```python
# 在没有数据扩充的情况下
#history = model.fit(X_train, Y_train, batch_size = batch_size, epochs = epochs, 
#          validation_data = (X_val, Y_val), verbose = 2)
```

```python
# 通过数据增强来防止过拟合
datagen = ImageDataGenerator(
        featurewise_center=False,  # 对输入的图片每个通道减去每个通道对应均值
        samplewise_center=False,  # 每张图片减去样本均值, 使得每个样本均值为0
        featurewise_std_normalization=False,  # 将输入除以数据集的标准差
        samplewise_std_normalization=False,  # 将样本输入除以数据集的标准差
        zca_whitening=False,  # 去除样本之间的相关性
        rotation_range=10,  # 随机旋转图像（0到180度） 
        zoom_range = 0.1, # 随机缩放图像
        width_shift_range=0.1,   # 水平随机移动图像（总宽度的一部分）
        height_shift_range=0.1,  # 垂直随机移动图像（总高度的一部分）
        horizontal_flip=False,  # 水平随机翻转图像
        vertical_flip=False)  # 垂直随机翻转图像

datagen.fit(X_train)
```

```python
# 拟合模型
epochs = 30 # epochs=30能得到0.987+的准确度
batch_size = 86
history = model.fit_generator(datagen.flow(X_train, Y_train, batch_size=batch_size),
                              epochs=epochs, validation_data=(X_val, Y_val),
                              verbose=2, steps_per_epoch=X_train.shape[0] // batch_size, callbacks=[learning_rate_reduction])
```

### 混淆矩阵（Confusion matrix）

&emsp;&emsp;混淆矩阵有助于发现模型的缺陷，绘制混淆矩阵如下图，可以看到：考虑到验证集（4200 张图像）的大小，我们的 CNN 在所有数字上的表现都非常好，几乎没有错误。但是，我们也注意到， CNN 在分类数字 4 的时候，被误分类为 9 的比较多，比较难以捕捉到 4 和 9 之间的差异。

```python
# 定义自己的混淆矩阵函数
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    - cm : 计算出的混淆矩阵的值
    - classes : 混淆矩阵中每一行每一列对应的列
    - normalize : True:显示百分比, False:显示个数
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    # plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# 绘制混淆矩阵
Y_pred = model.predict(X_val)    # 对验证数据集进行预测
Y_pred_classes = np.argmax(Y_pred, axis=1)  # 将预测的类转换为一个热向量
Y_true = np.argmax(Y_val, axis=1)  # 将验证集中实际观察到的类转换为一个热向量
confusion_mtx = confusion_matrix(Y_true, Y_pred_classes)  # 计算混淆矩阵

plt.figure(figsize=(16, 6))
plt.subplot(121)
plot_confusion_matrix(confusion_mtx, classes=range(10))
plt.subplot(122)
plot_confusion_matrix(confusion_mtx, classes=range(10), normalize=True)
```

&emsp;&emsp;我们来调查一下错误。我想看看最重要的错误。为此，我需要得到结果中实际值概率与预测值概率之间的差异。

```python
# 展示一些错误的结果
errors = (Y_pred_classes - Y_true != 0)  # 错误是指预测标签和真实标签之间的差异
Y_pred_classes_errors = Y_pred_classes[errors]
Y_pred_errors = Y_pred[errors]
Y_true_errors = Y_true[errors]
X_val_errors = X_val[errors]

def display_errors(errors_index, img_errors, pred_errors, obs_errors):
    """ 此函数显示 6 张图像及其预测标签和真实标签"""
    n = 0
    nrows = 2
    ncols = 3
    fig, ax = plt.subplots(nrows, ncols, figsize=(12, 6), sharex=True, sharey=True)
    for row in range(nrows):
        for col in range(ncols):
            error = errors_index[n]
            ax[row, col].imshow((img_errors[error]).reshape((28, 28)), cmap='gray')
            ax[row, col].set_title("Predicted label :{}\nTrue label :{}".format(pred_errors[error], obs_errors[error]))
            n += 1

Y_pred_errors_prob = np.max(Y_pred_errors, axis=1)  # 错误预测数字的概率
true_prob_errors = np.diagonal(np.take(Y_pred_errors, Y_true_errors, axis=1)) # 错误集中真实值的预测概率
delta_pred_true_errors = Y_pred_errors_prob - true_prob_errors  # 预测标签与真实标签的概率差异

sorted_dela_errors = np.argsort(-delta_pred_true_errors)  # 对数组进行降序序排序得到的索引
print('预测错误的共有：{}个\n预测标签与真实标签的概率差异为：\n{}'.format(len(Y_true_errors),
                                                  delta_pred_true_errors[sorted_dela_errors]))
most_important_errors = sorted_dela_errors[: 6]  
display_errors(most_important_errors, X_val_errors, Y_pred_classes_errors, Y_true_errors) # 展示前6个错误
```

对于这六个图像，该模型预测结果并不离谱。 因为其中一些错误也可能是人为的，比如说这个 4 十分地接近于 9，而这个 9 又似乎是 0，还有这个 3 特别像 8 。
