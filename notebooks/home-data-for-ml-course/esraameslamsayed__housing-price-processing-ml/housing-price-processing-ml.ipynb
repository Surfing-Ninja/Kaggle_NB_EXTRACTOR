# Building The Model  üíª 

# Feature and Target Selection

```python
y = df["SalePrice"]
X = df[['YearBuilt','LotArea','OverallQual','OverallCond','GrLivArea','1stFlrSF','2ndFlrSF','BedroomAbvGr','OpenPorchSF','PoolArea','SalePrice']]
X
```

# Normalizing the Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
```

# Spliting the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled,test_size=.2,random_state=1)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model_R = RandomForestRegressor(n_estimators=1000, random_state=42)
```

# Fitting the model

```python
model_R.fit(X_train,y_train)
```

# Predicting X_test

```python
y_pred=model_R.predict(X_test)
print(y_pred[:10])
```

```python
result_df = pd.DataFrame({'Actual': y_test.flatten() , 'Predicted': y_pred.flatten() })
result_df
```

```python
sns.lineplot(data=result_df, markers=True)
plt.xlabel('Observation (Index)')
plt.ylabel('Values')
plt.xlim(0,400)
plt.title('Line Plot of Actual vs. Predicted Values')
plt.show()
```

# Checking Accuracy of the Model

```python
from sklearn.metrics import mean_squared_error ,mean_absolute_error,r2_score
```

```python
acc = mean_absolute_error(y_pred,y_test)
print("Mean Absolute Error is: ",acc )
```


The r2_score function in scikit-learn
<br>calculates the R-squared (coefficient of determination) score, which is a measure of how well the predictions made by a regression model match the actual data. R-squared is a statistical measure that represents the proportion of the variance in the dependent variable (target) that is explained by the independent variables (features) in the model. It is often used to evaluate the goodness of fit of a regression model.
<br><i><br>
The R-squared score ranges from 0 to 1, where:
<br>
0 indicates that the model does not explain any variability in the target variable.
<br>
1 indicates that the model perfectly explains the variability in the target variable.

```python
score=r2_score(y_pred,y_test)
print("R-squared (coefficient of determination) score is : {:.2f}%".format(score * 100))
```
