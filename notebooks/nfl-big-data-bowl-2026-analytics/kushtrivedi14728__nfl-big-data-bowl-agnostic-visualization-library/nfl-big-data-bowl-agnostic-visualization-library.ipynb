# NFL Big Data Bowl: Agnostic Visualization Library

I’ve competed in the NFL Big Data Bowl for the past five years and developed a solid understanding of how the tracking data is structured. I’m planing to build an **agnostic visualization library** in plain ***Python*** and ***Matplotlib*** with *pip-installable* and prolly it may be useful to extend for each year’s competition goal. 

Since many elements stay consistent. This will be a first version; my goal is to release it by the end of the competition or before deadline (no promises, but I’ll try) while I focus on this year’s task. The aim is to make it available to everyone so they can ramp up faster on the competition goals in future
