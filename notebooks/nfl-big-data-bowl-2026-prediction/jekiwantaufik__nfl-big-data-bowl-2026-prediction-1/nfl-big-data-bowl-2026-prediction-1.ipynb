```python
import polars as pl

train_input_files = [
    f"/kaggle/input/nfl-big-data-bowl-2026-prediction/train/input_2023_w{str(i).zfill(2)}.csv"
    for i in range(1,19)
]

train_output_files = [
    f"/kaggle/input/nfl-big-data-bowl-2026-prediction/train/output_2023_w{str(i).zfill(2)}.csv"
    for i in range(1,19)
]

def load_concat_files(file_list):
    dfs = [pl.read_csv(f) for f in file_list]
    return pl.concat(dfs)

train_input = load_concat_files(train_input_files)
train_output = load_concat_files(train_output_files)

train_all = train_input.join(train_output, on=["game_id","play_id","nfl_id","frame_id"], how="inner")

train_all.write_csv("train_all.csv")
```
