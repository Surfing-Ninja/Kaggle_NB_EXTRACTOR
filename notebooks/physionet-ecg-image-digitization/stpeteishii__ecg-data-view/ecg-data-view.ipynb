```python
import os, glob, cv2, math, warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc
from PIL import Image 
from IPython.display import display

TRAIN_DIR    = '/kaggle/input/physionet-ecg-image-digitization/train'
TRAIN_CSV    = '/kaggle/input/physionet-ecg-image-digitization/train.csv'
```

```python
train=pd.read_csv(TRAIN_CSV)
display(train)
```

```python
paths=[]
for dirname, _, filenames in os.walk(f'{TRAIN_DIR}/7663343/'):
    for filename in filenames:
        paths+=[(os.path.join(dirname, filename))]
print(paths)
```

```python
df=pd.read_csv(f'{TRAIN_DIR}/7663343/7663343.csv')
display(df)
display(df.info())
cols=df.columns.tolist()

gc.collect()
```

```python
for col in cols:
    plt.plot(df.index, df[col], marker='o')
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.title(f"{col}")
    plt.grid(True)
    plt.show()
```

```python
for path in paths:
    if path.endswith('.png'):
        print(path.split('/')[-1])
        img = Image.open(path)
        w, h = img.size
        img_resized = img.resize((w // 2, h // 2))
        display(img_resized)
        print()

gc.collect()
```
