```python
# import numpy as np
# from PIL import Image
# image_data=np.load("/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks/1008.npy")
# print(image_data.shape)
# print(image_data.dtype)

# import matplotlib.pyplot as plt 
# #from PIL import¬†Image
# plt.imshow(image_data,cmap="gray")
# plt.title("Image Loaded")
# plt.show()

# pil_image=Image.fromarray(image_data)
# pil_image.show()





# import numpy as np
# import matplotlib.pyplot as plt
# from PIL import Image

# # Step 1: Load the .npy file
# image_path = "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks/10070.npy"
# image_array = np.load(image_path)

# # Step 2: Print basic info
# print("Array shape:", image_array.shape)
# print("Array data type:", image_array.dtype)

# # Step 3: Display using matplotlib
# plt.imshow(image_array, cmap='gray')
# plt.title("Image Loaded from .npy file")
# plt.axis('off')
# plt.show()

# # Step 4: Convert to uint8 if needed (for PIL)
# if image_array.dtype != np.uint8:
#     # Normalize to 0‚Äì255 range for visualization
#     normalized = (image_array - image_array.min()) / (image_array.max() - image_array.min())
#     image_array = (normalized * 255).astype(np.uint8)

# # Step 5: Display using PIL
# image = Image.fromarray(image_array)
# image.show()

# # Optional: Save the image
# # image.save("output_image.png")




# import numpy as np
# import matplotlib.pyplot as plt
# from PIL import Image

# # Step 1: Load the .npy file
# image_path = "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks/10070.npy"
# image_array = np.load(image_path)

# # Step 2: Remove any extra dimensions (like 1-channel)
# image_array = np.squeeze(image_array)

# print("Array shape after squeeze:", image_array.shape)
# print("Array dtype:", image_array.dtype)

# # Step 3: Display using matplotlib
# plt.imshow(image_array, cmap='gray')
# plt.title("Image Loaded from .npy file")
# plt.axis('off')
# plt.show()

# # Step 4: Convert to uint8 for PIL if needed
# if image_array.dtype != np.uint8:
#     normalized = (image_array - image_array.min()) / (image_array.max() - image_array.min())
#     image_array = (normalized * 255).astype(np.uint8)

# # Step 5: Display using PIL
# image = Image.fromarray(image_array)
# image.show()






# import numpy as np
# import matplotlib.pyplot as plt
# from PIL import Image

# # Load the .npy file
# image_path = "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks/10636.npy"
# image_array = np.load(image_path)
# print("Original shape:", image_array.shape)

# # If shape has more than 2 dimensions, handle appropriately
# if image_array.ndim == 3 and image_array.shape[0] <= 4:
#     # Move channels to the last dimension if needed (2, H, W) ‚Üí (H, W, 2)
#     image_array = np.transpose(image_array, (1, 2, 0))
#     print("Transposed shape:", image_array.shape)

# # Normalize and convert to uint8
# normalized = (image_array - image_array.min()) / (image_array.max() - image_array.min())
# image_uint8 = (normalized * 255).astype(np.uint8)

# # Display each channel separately
# for i in range(image_uint8.shape[-1]):
#     plt.imshow(image_uint8[..., i], cmap='gray')
#     plt.title(f"Channel {i+1}")
#     plt.axis('off')
#     plt.show()
# # Convert (H, W, 2) to (H, W, 3) by adding a third zero channel
# if image_uint8.shape[-1] == 2:
#     h, w, _ = image_uint8.shape
#     merged = np.zeros((h, w, 3), dtype=np.uint8)
#     merged[..., :2] = image_uint8  # fill R and G
#     plt.imshow(merged)
#     plt.title("Merged 2-Channel Image (Fake RGB)")
#     plt.axis('off')
#     plt.show()



import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# ================================
# 1. Load the .npy file
# ================================
image_path = "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks/10632.npy"
image_array = np.load(image_path)
print("Original shape:", image_array.shape)

# ================================
# 2. Normalize and convert to uint8
# ================================
image_array = image_array.astype(np.float32)
image_array = (image_array - image_array.min()) / (image_array.max() - image_array.min() + 1e-8)
image_uint8 = (image_array * 255).astype(np.uint8)

# ================================
# 3. Adjust shape for visualization
# ================================
if image_uint8.ndim == 3 and image_uint8.shape[0] in [1, 2, 3, 4]:
    # If channel-first (C, H, W) -> (H, W, C)
    image_uint8 = np.transpose(image_uint8, (1, 2, 0))

print("Adjusted shape:", image_uint8.shape)

# ================================
# 4. Display logic
# ================================
if image_uint8.ndim == 2:
    # Grayscale
    plt.imshow(image_uint8, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()

elif image_uint8.ndim == 3:
    channels = image_uint8.shape[2]
    if channels == 1:
        # Single-channel grayscale
        plt.imshow(image_uint8.squeeze(), cmap='gray')
        plt.title("Single-Channel Image")
        plt.axis('off')
        plt.show()

    elif channels == 2:
        # Two-channel image ‚Äî show each channel separately
        for i in range(2):
            plt.imshow(image_uint8[..., i], cmap='gray')
            plt.title(f"Channel {i+1}")
            plt.axis('off')
            plt.show()
        # Optionally merge as fake RGB for preview
        merged = np.zeros((image_uint8.shape[0], image_uint8.shape[1], 3), dtype=np.uint8)
        merged[..., :2] = image_uint8
        plt.imshow(merged)
        plt.title("Merged 2-Channel (Fake RGB)")
        plt.axis('off')
        plt.show()

    elif channels in [3, 4]:
        # RGB or RGBA
        plt.imshow(image_uint8)
        plt.title(f"{'RGB' if channels==3 else 'RGBA'} Image")
        plt.axis('off')
        plt.show()

    else:
        print(f"Unsupported number of channels: {channels}")

else:
    print("Unsupported image shape:", image_uint8.shape)

# ================================
# 5. Optional: Save one channel or full image
# ================================
# Example: save first channel
# Image.fromarray(image_uint8[..., 0]).save("output_channel1.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

def explore_segmentation_data(image_dir, mask_dir):
    """Explore segmentation dataset"""
    print("=" * 60)
    print("SEGMENTATION DATASET EXPLORATION")
    print("=" * 60)
    
    image_paths = sorted(list(Path(image_dir).glob('*.png')) + 
                        list(Path(image_dir).glob('*.png')))
    mask_paths = sorted(list(Path(mask_dir).glob('*.npy')))
    
    print(f"\nüìä Dataset Statistics:")
    print(f"  Images: {len(image_paths)}")
    print(f"  Masks: {len(mask_paths)}")
    
    # Check first sample
    if len(image_paths) > 0 and len(mask_paths) > 0:
        img = np.array(Image.open(image_paths[0]))
        mask = np.load(mask_paths[0])
        
        print(f"\nüìê Image Properties:")
        print(f"  Shape: {img.shape}")
        print(f"  Dtype: {img.dtype}")
        print(f"  Range: [{img.min()}, {img.max()}]")
        
        print(f"\nüé≠ Mask Properties:")
        print(f"  Shape: {mask.shape}")
        print(f"  Dtype: {mask.dtype}")
        print(f"  Unique values: {np.unique(mask)}")
        print(f"  Range: [{mask.min()}, {mask.max()}]")
        
        # Calculate forgery percentage
        if len(mask.shape) == 3:
            mask = mask[:, :, 0]  # Take first channel if RGB
        
        forged_pixels = (mask > 0).sum()
        total_pixels = mask.size
        forgery_ratio = forged_pixels / total_pixels * 100
        
        print(f"\nüîç Forgery Statistics:")
        print(f"  Forged pixels: {forged_pixels:,} ({forgery_ratio:.2f}%)")
        print(f"  Authentic pixels: {total_pixels - forged_pixels:,} ({100-forgery_ratio:.2f}%)")
        
        # Visualize
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        axes[0].imshow(img)
        axes[0].set_title('Original Image')
        axes[0].axis('off')
        
        axes[1].imshow(mask, cmap='gray')
        axes[1].set_title('Ground Truth Mask')
        axes[1].axis('off')
        
        # Overlay
        overlay = img.copy()
        if len(overlay.shape) == 2:
            overlay = np.stack([overlay]*3, axis=-1)
        overlay[mask > 0] = [255, 0, 0]  # Red for forged regions
        
        axes[2].imshow(overlay)
        axes[2].set_title('Overlay (Red = Forged)')
        axes[2].axis('off')
        
        plt.tight_layout()
        plt.savefig('results/data_exploration_segmentation.png', dpi=100)
        plt.show()
        
    return image_paths, mask_paths

# Run exploration
image_paths, mask_paths = explore_segmentation_data(
    "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_images/authentic",
    "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks",
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

def explore_segmentation_data(image_dir, forged_dir, mask_dir):
    """Explore segmentation dataset with authentic + forged images"""
    print("=" * 60)
    print("SEGMENTATION DATASET EXPLORATION")
    print("=" * 60)
    
    image_paths = sorted(list(Path(image_dir).glob('*.png')))
    forged_paths = sorted(list(Path(forged_dir).glob('*.png')))
    mask_paths = sorted(list(Path(mask_dir).glob('*.npy')))
    
    print(f"\nüìä Dataset Statistics:")
    print(f"  Authentic Images: {len(image_paths)}")
    print(f"  Forged Images: {len(forged_paths)}")
    print(f"  Masks: {len(mask_paths)}")
    
    # Check first sample
    if len(image_paths) > 0 and len(mask_paths) > 0 and len(forged_paths) > 0:
        # Load data
        img_auth = np.array(Image.open(image_paths[0]))
        img_forged = np.array(Image.open(forged_paths[0]))
        mask = np.load(mask_paths[0])
        
        print(f"\nüìê Image Properties:")
        print(f"  Authentic Image: {img_auth.shape}, dtype={img_auth.dtype}")
        print(f"  Forged Image: {img_forged.shape}, dtype={img_forged.dtype}")
        
        print(f"\nüé≠ Mask Properties:")
        print(f"  Shape: {mask.shape}, dtype={mask.dtype}")
        print(f"  Unique values: {np.unique(mask)}")
        
        # ‚úÖ Clean mask
        mask = np.squeeze(mask)
        if mask.ndim == 3 and mask.shape[0] in [2, 3, 4]:
            mask = np.transpose(mask, (1, 2, 0))
            mask = mask[..., 0]
        
        # ‚úÖ Resize mask if needed
        if mask.shape[:2] != img_auth.shape[:2]:
            print(f"‚ö†Ô∏è Resizing mask from {mask.shape} to {img_auth.shape[:2]}")
            mask = np.array(Image.fromarray(mask).resize(img_auth.shape[:2][::-1]))
        
        # Forgery statistics
        forged_pixels = (mask > 0).sum()
        total_pixels = mask.size
        forgery_ratio = forged_pixels / total_pixels * 100
        
        print(f"\nüîç Forgery Statistics:")
        print(f"  Forged pixels: {forged_pixels:,} ({forgery_ratio:.2f}%)")
        
        # ‚úÖ Overlay creation
        overlay = img_auth.copy()
        if len(overlay.shape) == 2:
            overlay = np.stack([overlay]*3, axis=-1)
        overlay[mask > 0] = [255, 0, 0]  # red overlay
        
        # ===============================
        # üî∏ Visualization (4 images)
        # ===============================
        fig, axes = plt.subplots(1, 4, figsize=(20, 5))
        
        axes[0].imshow(img_auth)
        axes[0].set_title('Authentic Image')
        axes[0].axis('off')
        
        axes[1].imshow(mask, cmap='gray')
        axes[1].set_title('Ground Truth Mask')
        axes[1].axis('off')
        
        axes[2].imshow(overlay)
        axes[2].set_title('Overlay (Red = Forged)')
        axes[2].axis('off')

        axes[3].imshow(img_forged)
        axes[3].set_title('Forged Image')
        axes[3].axis('off')
        
        plt.tight_layout()
        Path("results").mkdir(exist_ok=True)
        plt.savefig('results/data_exploration_segmentation.png', dpi=100)
        plt.show()
        
    else:
        print("‚ùå Missing images or masks for exploration.")
    
    return image_paths, forged_paths, mask_paths
# Run exploration
image_paths, forged_paths, mask_paths = explore_segmentation_data(
    "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_images/authentic",
    "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_images/forged",
    "/kaggle/input/recodai-luc-scientific-image-forgery-detection/train_masks",
)
```
