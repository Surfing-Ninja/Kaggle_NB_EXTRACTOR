```python
samples = os.listdir('/kaggle/input/recodai-luc-scientific-image-forgery-detection/test_images/')
samples[:5]
```

```python
case_id = []
annotation = []
for sample in samples:
    case_id.append(sample.split('.')[0])
    if 1:#np.random.rand() < .125:
        annotation.append('authentic')
    else:
        annotation.append(
            rle_encode(
                list((np.random.rand(*np.load(src_pth + masks[np.random.randint(len(masks))]).shape) < .5).astype(np.uint8))
            )
        )
```

```python
submission = pd.DataFrame({
    'case_id':case_id,
    'annotation':annotation
})
submission.head(5)
```

```python
submission.to_csv('submission.csv',index=False)
```
