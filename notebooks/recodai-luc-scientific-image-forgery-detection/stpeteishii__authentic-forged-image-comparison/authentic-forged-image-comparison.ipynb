# **Authentic/Forgery Image Comparison**

```python
import os, glob, cv2, math, warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def show(i):
    path='/kaggle/input/recodai-luc-scientific-image-forgery-detection'
    img1 = mpimg.imread(f'{path}/train_images/authentic/{i}.png')
    img2 = mpimg.imread(f'{path}/train_images/forged/{i}.png')
    img3 = np.load(f'{path}/train_masks/{i}.npy')[0]
    titles=['authentic','forged','mask']
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))  
    for ax, img, title in zip(axes, [img1, img2, img3], titles):
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(title)
    plt.tight_layout()
    plt.show()
```
