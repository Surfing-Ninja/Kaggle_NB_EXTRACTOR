### A dummy submission solution that predicts all test samples as “authentic”

```python
import os
import pandas as pd
from glob import glob

DATA_PATH = DATA_PATH="/kaggle/input/recodai-luc-scientific-image-forgery-detection/" 

# ==================== CREATE DUMMY SUBMISSION ====================
test_image_paths = sorted(glob(os.path.join(DATA_PATH, "test_images", "*.png")))

submission_data = [
    {"case_id": os.path.splitext(os.path.basename(p))[0], "annotation": "authentic"}
    for p in test_image_paths
]

submission_df = pd.DataFrame(submission_data)
submission_df.to_csv("submission.csv", index=False)

print("Submission created!")
print(f"Total test images: {len(submission_df)}")
print(submission_df.head())
```
